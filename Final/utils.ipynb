```python
## Imports
import shutil
from pathlib import Path

import pandas as pd
```

```python
# Const declarations
train_csv_path="../train.csv"
test_csv_path="../test.csv"
dataset_dir="../Datasets/Originals"

train_dir_path="../Datasets/Train"
val_dir_path="../Datasets/Validation"
test_dir_path="../Datasets/Test"

train_cancer_dir_path = "../Datasets/Train/cancer"
train_normal_dir_path = "../Datasets/Train/normal"
val_cancer_dir_path = "../Datasets/Validation/cancer"
val_normal_dir_path = "../Datasets/Validation/normal"
test_cancer_dir_path = "../Datasets/Test/cancer"
test_normal_dir_path = "../Datasets/Test/normal"


# images props
img_height = 512
img_width = 256
```

```python
## Functions
def build_image_path(image_id: str) -> str:
    return Path(dataset_dir, str(image_id) + '.png').resolve()

def clean_all_datasets() -> None:
    # clean folder before augmentation
    shutil.rmtree(train_cancer_dir_path)
    shutil.rmtree(train_normal_dir_path)
    shutil.rmtree(val_cancer_dir_path)
    shutil.rmtree(val_normal_dir_path)
    shutil.rmtree(test_cancer_dir_path)
    shutil.rmtree(test_normal_dir_path)

def create_dataset_dirs() -> None:
    Path(train_cancer_dir_path).mkdir(parents=True, exist_ok=True)
    Path(train_normal_dir_path).mkdir(parents=True, exist_ok=True)
    Path(val_cancer_dir_path).mkdir(parents=True, exist_ok=True)
    Path(val_normal_dir_path).mkdir(parents=True, exist_ok=True)
    Path(test_cancer_dir_path).mkdir(parents=True, exist_ok=True)
    Path(test_normal_dir_path).mkdir(parents=True, exist_ok=True)

def build_train_dataset_from_ori_imgs(df: pd.DataFrame) -> None:
  for path in df[df["cancer"]==1]['img_path']:
      shutil.copy2(path, train_cancer_dir_path)

  for path in df[df["cancer"]==0]['img_path']:
      shutil.copy2(path, train_normal_dir_path)


def build_val_dataset_from_ori_imgs(df: pd.DataFrame) -> None:
  for path in df[df["cancer"]==1]['img_path']:
      shutil.copy2(path, val_cancer_dir_path)
      
  for path in df[df["cancer"]==0]['img_path']:
      shutil.copy2(path, val_normal_dir_path)

def build_test_dataset_from_ori_imgs(df: pd.DataFrame) -> None:
  for path in df[df["cancer"]==1]['img_path']:
      shutil.copy2(path, test_cancer_dir_path)
      
  for path in df[df["cancer"]==0]['img_path']:
      shutil.copy2(path, test_normal_dir_path)

def print_img_props(image):
    print(f"Image props -> Shape: {image.shape}, MaxVal: {image.max()}, MinVal: {image.min()}, Type: {image.dtype}")
```
